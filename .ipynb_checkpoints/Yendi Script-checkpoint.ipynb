```python
import geopandas as gpd
from shapely.geometry import box
from shapely.ops import unary_union
import math
import os
import matplotlib.pyplot as plt
import random

def create_grid(polygon, cell_area, crs):
    """
    Create a grid of squares (or rectangles) covering the polygon.
    
    Parameters:
    polygon (shapely.geometry.Polygon): The input polygon to cover.
    cell_area (float): The area of each cell in the grid in square meters (or other units).
    crs (str or dict): The CRS to assign to the grid.
    
    Returns:
    geopandas.GeoDataFrame: A GeoDataFrame containing the grid cells as polygons.
    """
    minx, miny, maxx, maxy = polygon.bounds
    
    # Calculate the cell size (assuming square cells for simplicity)
    cell_size = math.sqrt(cell_area)
    
    # Debugging: Print bounds and cell size
    print(f"Polygon bounds: {polygon.bounds}")
    print(f"Calculated cell size: {cell_size}")
    
    # Create the grid cells
    cells = []
    x = minx
    while x < maxx:
        y = miny
        while y < maxy:
            cell = box(x, y, x + cell_size, y + cell_size)
            cells.append(cell)
            y += cell_size
        x += cell_size

    # Debugging: Print the number of cells created
    print(f"Number of cells created: {len(cells)}")

    # Create a GeoDataFrame from the grid cells
    grid = gpd.GeoDataFrame({'geometry': cells}, crs=crs)
    
    return grid

def generate_random_values():
    """
    Generate random values for properties while maintaining specific relationships.
    
    Returns:
    tuple: A tuple containing (prop_value, income_level, pop_density, agric_index)
    """
    pop_density = random.uniform(100.0, 5000.0)
    agric_index = 1.0 - (pop_density / 5000.0)  # Inverse relationship with pop_density
    income_level = random.uniform(500.0, 11000.0)
    prop_value = income_level * (pop_density / 5000.0) * 10  # Direct relationship with income_level and pop_density
    jurisdiction = random.uniform(1450.0, 1750.0)
    distance_to_water = random.uniform(100, 1750.0)
    enforcement_effectiveness = random.uniform(0.0, 1.0)

    
    return prop_value, income_level, pop_density, agric_index, jurisdiction, distance_to_water, enforcement_effectiveness

def clip_grid_to_polygon(grid, polygon):
    """
    Clip the grid to the input polygon.
    
    Parameters:
    grid (geopandas.GeoDataFrame): The input grid of cells.
    polygon (shapely.geometry.Polygon): The polygon to clip the grid to.
    
    Returns:
    geopandas.GeoDataFrame: A GeoDataFrame containing the clipped grid cells.
    """
    # Perform the clipping
    clipped_grid = grid[grid.intersects(polygon)].copy()
    clipped_grid['geometry'] = clipped_grid.intersection(polygon)
    
    # Assign random values to each row based on the relationships
    prop_values, income_levels, pop_densities, agric_indices, jurisdiction, distance_to_water, enforcement_effectiveness = [], [], [], [],[], [], []
    for _ in range(len(clipped_grid)):
        prop_value, income_level, pop_density, agric_index, jurisdiction_index, distance_to_water_index, enforcement_effectiveness_index = generate_random_values()
        prop_values.append(prop_value)
        income_levels.append(income_level)
        pop_densities.append(pop_density)
        agric_indices.append(agric_index)
        jurisdiction.append(jurisdiction_index)
        distance_to_water.append(distance_to_water_index)
        enforcement_effectiveness.append(enforcement_effectiveness_index)
        
    
    clipped_grid['prop_value'] = prop_values
    clipped_grid['income_level'] = income_levels
    clipped_grid['pop_density'] = pop_densities
    clipped_grid['agric_index'] = agric_indices
    clipped_grid['jurisdiction'] = jurisdiction
    clipped_grid['distance_to_water'] = distance_to_water
    clipped_grid['enforcement_effectiveness'] = enforcement_effectiveness
    
    # Filter out any empty geometries that might have resulted from clipping
    clipped_grid = clipped_grid[~clipped_grid.is_empty]
    
    return clipped_grid

# Define the path to your input polygon file
input_polygon_path = aoi

# Define the desired area of each sub-polygon in square meters
desired_sub_polygon_area = 1000000  # 1 hectare (10000 square meters)

# Read the input polygon
gdf = gpd.read_file(input_polygon_path)

# Ensure that there are valid geometries in the GeoDataFrame
if gdf.empty or gdf.geometry.isna().all():
    raise ValueError("The input shapefile does not contain valid geometries.")

# Convert the GeoDataFrame to a projected CRS for accurate area and distance calculations
gdf_projected = gdf.to_crs(epsg=32630)  # UTM Zone 30N for Ghana

# Perform the unary union of all geometries to get a single polygon
input_polygon = gdf_projected.geometry.unary_union

if input_polygon.is_empty:
    raise ValueError("The resulting union of geometries is empty.")

# Create the grid
grid = create_grid(input_polygon, desired_sub_polygon_area, gdf_projected.crs)

# Clip the grid to the input polygon
clipped_grid = clip_grid_to_polygon(grid, input_polygon)

# Debugging: Check the number of cells after clipping
print(f"Number of clipped grid cells: {len(clipped_grid)}")

# Convert the clipped grid back to the original geographic CRS
clipped_grid = clipped_grid.to_crs(epsg=4326)

# Save the clipped grid to a new shapefile
output_shapefile_path = os.path.join(os.path.dirname(input_polygon_path), 'output_sub_polygons.shp')
clipped_grid.to_file(output_shapefile_path)

print(f"Sub-polygons shapefile saved to {output_shapefile_path}")

# Plot the input polygon and the grid
fig, ax = plt.subplots(figsize=(10, 10))
gdf.to_crs(epsg=4326).boundary.plot(ax=ax, edgecolor='black', linewidth=1, label='Original Polygon')
grid.to_crs(epsg=4326).boundary.plot(ax=ax, edgecolor='blue', linestyle='--', linewidth=0.5, label='Grid Cells')  # Plot the entire grid
clipped_grid.to_crs(epsg=4326).boundary.plot(ax=ax, edgecolor='red', linewidth=1, label='Clipped Grid Cells')  # Plot the clipped grid

# Add legend and titles
ax.legend()
plt.title('Polygon Splitting into Sub-Polygons')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

### Overlay Analysis
Overlay economic, social, and regulatory data to identify high-risk areas.

```python
# Example overlay analysis to identify high-risk areas
high_threshold_prop_value = clipped_grid['prop_value'].quantile(0.75)
high_threshold_income_level = clipped_grid['income_level'].quantile(0.75)
high_threshold_pop_density = clipped_grid['pop_density'].quantile(0.75)

high_risk_areas = clipped_grid[(clipped_grid['prop_value'] > high_threshold_prop_value) & 
                               (clipped_grid['income_level'] > high_threshold_income_level) & 
                               (clipped_grid['pop_density'] > high_threshold_pop_density)]

fig, ax = plt.subplots(figsize=(10, 7))
high_risk_areas.plot(column='prop_value', legend=True, cmap='Reds', ax=ax)
ax.set_title('High-Risk Areas for Encroachment')

plt.tight_layout()
plt.show()

# Display the number of high-risk areas identified
display(high_risk_areas.shape[0])
```

### Hotspot Analysis
Use spatial statistical techniques to identify hotspots of encroachment activity.

```python
import esda
import pysal.lib as lib
from splot.esda import plot_moran

# Calculate Moran's I to identify hotspots
w = lib.weights.Queen.from_dataframe(clipped_grid)
w.transform = 'r'
moran = esda.Moran(clipped_grid['prop_value'], w)

plot_moran(moran)
plt.show()

# Display Moran's I and p-value
display(f"Moran's I: {moran.I}, p-value: {moran.p_sim}")
```

### Accessibility Analysis
Analyze changes in access to natural resources using network analysis and proximity tools.

```python
# Calculate distance to the nearest waterway
water= r"C:\Users\Justine Kojo\Desktop\Projects\GIS\Other Agencies\Desmond\August\Yendi - Encroachment on public lands\Spatial\waterways.gpkg"
waterways = gpd.read_file(water).to_crs(epsg=25000)
clipped_grid['distance_to_water'] = clipped_grid.to_crs(epsg=25000).geometry.apply(
    lambda x: waterways.distance(x).min())

# Plot distance to waterway
fig, ax = plt.subplots(figsize=(10, 7))

clipped_grid.plot(column='distance_to_water', legend=True, cmap='Blues_r', scheme='quantiles', ax=ax)
ax.set_title('Distance to Nearest Waterway')

plt.tight_layout()
plt.show()

# Display the summary statistics for distance to waterway
display(clipped_grid['distance_to_water'].describe())
```
